## Test the loading of a csv file

```python
from civiclink.utils import get_root_dir
import pandas as pd

root_dir = get_root_dir()
fpath = root_dir/"data/fake_steuerdaten_labels_not_random.csv"
df = pd.read_csv(fpath, delimiter=";")
```

```python
from xgboost import XGBClassifier

from sklearn.model_selection import train_test_split

X,y = df.drop(columns="Label"), df["Label"]

for col in ["Familienstand", "Steuerklasse", "Bundesland","Religionszugehörigkeit", 
            "Einkunftsart","Branche_Selbstaendig"]:
    if col in X.columns:
        X[col] = X[col].astype("category")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3)

bst = XGBClassifier(enable_categorical=True)

bst.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

y_pred = bst.predict_proba(X_test)
score = roc_auc_score(y_test,y_pred[:,1])

score
```

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(n_splits=10, random_state=42, shuffle=True)

scores = []
for i, (train_index, test_index) in enumerate(cv.split(X, y)):
    print(f"Fold {i}:")
    bst.fit(X.iloc[train_index], y.iloc[train_index])
    y_pred = bst.predict_proba(X.iloc[test_index])
    score = roc_auc_score(y.iloc[test_index],y_pred[:,1])
    print(score)
    scores.append(score)

print(sum(scores)/len(scores))


```
